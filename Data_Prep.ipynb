# MCS

```python
user = "/Users/Colette/anaconda_important_code/acapstone"
```

```python
mcs_data = pd.read_csv(user + "/data/data_prep/prep_MCS_fuel.csv")
```

```python
mcs_data
```

```python
mcs_data["date"] = pd.to_datetime(mcs_data[['year', 'month', 'day']])
```

```python
mcs_data.head()
```

```python
mcs_data["new_fuel"] = ""
```

```python
for row in range(len(mcs_data.month)-1):
    if mcs_data.month[row] != mcs_data.month[row+1]:
        total_days_in_month = calendar.monthrange(2017, mcs_data.month[row])[1]
        invoice_days = total_days_in_month - (mcs_data.day[row])
        invoice_days_next_month =  mcs_data.day[row+1]
        last_invoice = invoice_days * (2662.838 / (invoice_days + invoice_days_next_month))
        next_invoice = invoice_days_next_month * (2662.838 / (invoice_days + invoice_days_next_month))
        mcs_data.new_fuel[row] = last_invoice
        mcs_data.new_fuel[row + 1] = next_invoice 
        mcs_data.fuel[row] = 2662.838 + mcs_data.new_fuel[row]
        mcs_data.fuel[row + 1] =  mcs_data.new_fuel[row + 1]
    else:
        print('no')
        print(mcs_data.fuel[row])
```

```python
mcs_data.head()
```

```python
mcs_data = mcs_data.drop([0])
```

```python
mcs_data.head()
```

```python
aggregation_functions = {"year": "first", 
                         "fuel": "sum", 
                         "production": "first", 
                         "factory": "first", 
                         "type": "first", 
                         "supplier": "first", 
                         "unit_fuel": "first", 
                         "unit_production": "first"}
data_mcs = mcs_data.groupby(mcs_data["month"], as_index=False).aggregate(aggregation_functions)
```

```python
data_mcs
```

```python
data_mcs.to_csv(user + "/data/data_exploration/data_MCS.csv")
```

# TintColor

```python
tint_data = pd.read_csv(user + "/data/data_prep/prep_TintColor_fuel.csv")
```

```python
tint_data["date"] = pd.to_datetime(tint_data[['year', 'month', 'day']])

tint_data[:10]
```

```python
tint_data["new_fuel"] = ""
```

```python
tint_data.head()
```

```python
for row in range(len(tint_data.month)-1):
    if tint_data.month[row] != tint_data.month[row+1]:
        total_days_in_month = calendar.monthrange(2017, tint_data.month[row])[1]
        invoice_days = total_days_in_month - (tint_data.day[row])
        invoice_days_next_month =  tint_data.day[row+1]
        last_invoice = invoice_days * (tint_data.fuel[row] / (invoice_days + invoice_days_next_month))
        next_invoice = invoice_days_next_month * (tint_data.fuel[row] / (invoice_days + invoice_days_next_month))
        tint_data.new_fuel[row] = tint_data.fuel[row] + last_invoice
        tint_data.new_fuel[row + 1] =  next_invoice
        tint_data.fuel[row] = tint_data.new_fuel[row]
        tint_data.fuel[row + 1] = tint_data.new_fuel[row + 1]
    else:
        print('no')
```

```python
tint_data.head()
```

```python
aggregation_functions = {"year": "first", 
                         "fuel": "sum", 
                         "production": "first", 
                         "factory": "first", 
                         "type": "first", 
                         "supplier": "first", 
                         "unit_fuel": "first", 
                         "unit_production": "first"}
data_TintColor = tint_data.groupby(tint_data["month"], as_index=False).aggregate(aggregation_functions)
```
